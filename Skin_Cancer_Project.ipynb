```python
df = df.dropna()
```

```python
df
```

```python
new_categ = pd.get_dummies(df['localization'],prefix ='local')
```

```python
new_categ
```

```python
new_df = pd.concat([df,new_categ],axis=1)
```

```python
new_df
```

```python
new_categ_df = pd.get_dummies(df['dx_type'],prefix ='dx_type')
```

```python
new_categ_df
```

```python
new_data =  pd.concat([new_df,new_categ_df],axis=1)
```

```python
new_data
```

```python
new_data = new_data.drop(['localization','lesion_id','image_id','dx_type'],axis=1)
```

```python
new_data
```

```python
x = new_data.drop('dx',axis=1)
y = new_data['dx']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mylogmodel = LogisticRegression()
```

```python
mylogmodel.fit(x_train,y_train)
```

```python
skincancerpredictions = mylogmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,skincancerpredictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test,skincancerpredictions)
```

```python
accuracy
```
